```python
df2['Price'] = df2['Price'].astype('float')
```

```python
#A13.
average_prices = df2.groupby('Type')['Price'].mean().reset_index()

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(average_prices['Type'], average_prices['Price'], color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.show()
```
